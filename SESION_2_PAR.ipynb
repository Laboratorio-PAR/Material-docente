##### NOTA: Rangos en Python

##### Descripci√≥n
Un rango es una secuencia de n√∫meros que se genera de manera autom√°tica, √∫til para iterar sobre secuencias de manera eficiente.
Se crea utilizando la funci√≥n `range()`.

##### Sintaxis de `range()`

 - `range(stop)`: Genera n√∫meros desde `0` hasta `stop - 1`.
 - `range(start, stop)`: Genera n√∫meros desde `start` hasta `stop - 1`.
 - `range(start, stop, step)`: Genera n√∫meros desde `start` hasta `stop - 1` con saltos de tama√±o `step`.

##### Usos comunes
- Iterar sobre secuencias de n√∫meros.
- Crear listas r√°pidamente.
- Generar √≠ndices para estructuras como listas o matrices.

##### Ejemplos de uso:

````python
# #### Rango simple
for i in range(5):
    print("N√∫mero en el rango simple:", i)

# #### Rango con inicio y fin
for i in range(2, 6):
    print("N√∫mero en el rango con inicio y fin:", i)

# #### Rango con pasos
for i in range(0, 10, 2):
    print("N√∫mero en el rango con pasos:", i)

# #### Convertir un rango en una lista
rango_a_lista = list(range(5))
print("Rango convertido en lista:", rango_a_lista)
````

- En python las funciones son **first class citizens**, es decir, se pueden pasar como argumento a otras funciones, se pueden devolver como resultado de otras funciones, se pueden guardar en variables, etc.
- Como cualquier otro objeto en Python, se puede guardar una funci√≥n (o referencia) en una variable

````python
# procesar strings aplicando funciones en 'ops'
def clean_strings(strings, ops):
    result = []
    for value in strings:
        for function in ops:
            value = function(value)
        result.append(value)
    return result

strings = ['  valencia   ', '   barcelona', ' bilbao ']
operations = [str.strip, str.title]
print(clean_strings(strings,operations))
````

### Paso de Par√°metros
- Tipos simples por valor, int, float, string
 - Se crea una copia dentro de la funci√≥n
- Tipos complejos por referencia, list, set, dictionary
 - Se maneja directamente la variable externa, los cambios le afectan

````python
# paso por valor
def pow2(x):
    x *= x
    print(x)
    return

a = 2
pow2(a)
print(a)
````

````python
# paso por referencia
def add2(x):
    x.append(2)
    return

lista = [0, 1]
add2(lista)
print(lista)
````

````python
# paso por valor modificando, devolver el valor
def pow2(x):
    return x * x

a = 2
a = pow2(a)
print(a)
````

````python
# paso por referencia sin modificar, enviando copia
def add2(x):
    x.append(2)

lista = [0, 1]
add2(lista.copy())
print(lista)
````

````python
lista = [0,1]

def add2():
    lista.append(2)

add2()
print(lista)
````

### La Filosof√≠a de Python: Clases y PEPs

Antes de escribir c√≥digo, es vital entender dos aspectos clave de la filosof√≠a de Python:

1.  **"Somos todos adultos y responsables"**: A diferencia de Java/PHP, Python no tiene palabras clave `private`, `public` o `protected`. La encapsulaci√≥n se basa en *convenciones*. Confiamos en que el programador no acceder√° a partes internas de un objeto si no debe. Veremos c√≥mo la sintaxis `_` y `__` apoya esta idea.
2.  **PEP-8 y Nomenclatura**: La gu√≠a de estilo PEP-8 es clara: los nombres de las clases deben usar **`CamelCase`** (ej. `MiClase`), mientras que las variables y m√©todos deben usar **`snake_case`** (ej. `mi_metodo`).
3.  **Zen de Python**: La legibilidad y la simplicidad son claves. Si una implementaci√≥n es dif√≠cil de explicar, probablemente sea una mala idea.

### Definici√≥n de Clases, Instancias y Atributos

La sintaxis para definir una clase es simple. Usaremos un ejemplo pr√°ctico.

A continuaci√≥n se muestra un esquema de la sintaxis para definir una clase en Python:

```python
class <nombre> (<herencia>):
    '''Implementaci√≥n de la clase.'''
    def __init__(self):                     # Constructor: dos guiones bajos delante y atr√°s
                                            # self es el propio objeto

```

#### Atributos de instancia *vs* atributos de clase

Los atributos que se crean dentro del m√©todo ***__ init __*** se conocen como **atributos del objeto**, mientras que los que se crean fuera de √©l se conocen como **atributos de la clase**.
Mientras que los primeros son *propios de cada objeto* y por tanto pueden tomar valores distintos, los valores de los atributos de la clase *son los mismos para cualquier objeto de la clase*.

üìå En general, no deben usarse atributos de clase, excepto para almacenar valores constantes. üìå

````python
# 1. Definici√≥n de la clase
class Coche:
    """Una clase para modelar un coche (docstring de la clase)."""

    # 2. Atributo de Clase (compartido por todas las instancias)
    ruedas = 4

    # 3. El Constructor: __init__
    # 'self' es la referencia a la instancia (como 'this' en Java/PHP)
    def __init__(self, marca, modelo, anio):
        """Inicializa los atributos de la instancia."""
        # 4. Atributos de Instancia
        self.marca = marca
        self.modelo = modelo
        self.anio = anio
        self.kilometraje = 0  # Atributo con valor por defecto

    # 5. M√©todos de Instancia
    def get_descripcion(self):
        """Devuelve una descripci√≥n formateada del coche."""
        return f"{self.anio} {self.marca.title()} {self.modelo.title()}"

    def leer_kilometraje(self):
        """Muestra el kilometraje del coche."""
        print(f"Este coche tiene {self.kilometraje} km.")

    def actualizar_kilometraje(self, km):
        """Actualiza el kilometraje, evitando retrocesos."""
        if km >= self.kilometraje:
            self.kilometraje = km
        else:
            print("¬°No se puede retroceder el cuentakil√≥metros!")

    def incrementar_kilometraje(self, km):
        """Suma una cantidad al kilometraje total."""
        if km >= 0:
            self.kilometraje += km
````

### Instanciaci√≥n y Uso

Crear un objeto (*instanciar*) se ve igual que llamar a una funci√≥n. El m√©todo `__init__` se invoca autom√°ticamente.

Este m√©todo es un m√©todo especial que se llama cada vez que se instancia una clase y sirve para inicializar el objeto que se crea. Este m√©todo crea los atributos que deben tener todos los objetos de la clase y por tanto contiene los par√°metros necesarios para su creaci√≥n, pero no devuelve nada.
Se invoca cada vez que se instancia un objeto de esa clase.

````python
# Creamos una instancia de Coche
mi_coche = Coche('seat', 'ibiza', 2021)

# Accedemos a los atributos de instancia
print(mi_coche.get_descripcion())

# Accedemos al atributo de clase
print(f"Ruedas: {mi_coche.ruedas}") # A trav√©s de la instancia
print(f"Ruedas (desde la clase): {Coche.ruedas}") # A trav√©s de la clase

# Modificamos atributos (directamente o por m√©todos)
mi_coche.leer_kilometraje()
mi_coche.actualizar_kilometraje(100)
mi_coche.leer_kilometraje()

# Intentamos retroceder (fallar√°)
mi_coche.actualizar_kilometraje(50)

mi_coche.incrementar_kilometraje(25)
mi_coche.leer_kilometraje()
````
