### **Definición**
Una cadena es un conjunto de alfabetos, palabras u otros caracteres. Es una de las estructuras de datos primitivas y son los bloques de construcción para la manipulación de datos.

- Python tiene una clase de cadena incorporada llamada **str**.
- Las cadenas de Python son "inmutables", lo que significa que no se pueden modificar después de crearlas. (Son **TUPLAS**)
- Para manipular cadenas, creamos nuevas cadenas sobre la marcha para representar valores calculados, debido a su propiedad inmutable.

Como se ha comentado, es posible manejar datos textuales en Python mediante el objeto **str**.
Las cadenas son *secuencias inmutables de Unicode*. *Unicode* es un sistema diseñado para representar todos los caracteres de las lenguas. En *Unicode*, cada letra o carácter se representa como un número de 4 bytes. Cada número representa un carácter único.

Para representar una cadena, se debe *envolver entre comillas*. Existen múltiples formas de hacerlo:

- **Comillas simples**, como en este ejemplo: 'Las comillas simples permiten incrustar comillas "dobles" en la cadena.'
- **Comillas dobles**. Por ejemplo: "Las comillas dobles permiten incrustar comillas 'simples' en la cadena."
- **Comillas triples**, como en este ejemplo: """Comillas triples con comillas dobles""", """Comillas triples con comillas simples""".

La cadena entre comillas triples permite trabajar con cadenas de varias líneas, y todos los espacios en blanco asociados se incluirán en la cadena.

Como ya se ha citado, las cadenas son inmutables, lo que significa que si se intenta cambiar cualquier valor de una cadena, se producirá un error. Será, por tanto, necesario crear una nueva cadena para incorporar los cambios, como se muestra a continuación:

```python
triple_quote = '''Esta es una cadena con triples comillas usando comillas "simples".'''
triple_quote[35] = "'"
```

```python
triple_quote_new = triple_quote[0:56] + "'dobles'" + triple_quote[65:]
print(triple_quote_new)
```

Para conocer la longitud de una cadena se usa la función `len()`

```python
print(len(triple_quote_new))
```

### Comparación de cadenas de texto.

Nota: es `case sensitive`(difencia mayúsculas de minúsculas)

```python
print("cadena" == "cadena") # True
print("cadena" != "Cadena") # True
print("CADENA".lower() == "cadena".upper()) # False
```

### Split - División de elementos de una cadena.

```python
a = "AMPLI;451;PARAD;729"
b = a.split(';') # en este caso, el delimitador es el ';', aunque puede ser cualquier caracter
print(b) # ['AMPLI', '451', 'PARAD', '729'])
```

### Búsqueda de caracteres o tramos dentro de una cadena.

Se utiliza la función `find()`que devuelve el índice más bajo de la cadena donde se encuentra la subcadena. También se puede especificar el índice inicial y final dentro de la cadena donde se desea que se busque la subcadena. Devuelve -1 si no se encuentra la subcadena.

```python
print("cadena".find('e')) # 3
print("cadena".find('una')) # -1
print("cadena".find('dena')) # 2
print('i' in "cadena") # False
```

### Conteo de caracteres o tramos dentro de una cadena.
La función `count(substring)` cuenta cuántas veces aparece una subcadena en la cadena. También se puede especificar el índice de inicio y de fin de la cadena.

```python
str1 = 'I got you a cookie, do you like cookies?'
str2 = 'cookie'
print("Numero de veces que aparece la cadena '"+ str2 + "' en la cadena '" + str1 + "': ", str1.count(str2))
```

### Acceso a los elementos/caracteres de una cadena.

```python
print("cadena"[0]) # c
print("cadena"[-1]) # a. Acceso desde el final de la cadena.
print("cadena"[-2]) # n. Acceso desde el final de la cadena.
```

### Concatenación de cadenas.

```python
print("cadena" + "s largas") # cadenas largas
```

### Slicing de cadenas.

```python
print("cadena"[0:]) # cadena
print("cadena"[:-2]) # cade
print("cadena"[2:-2]) # de
# Empieza:Acaba-1:Número_saltos
print("cadena muy larga"[2:-2:3]) # daul --> empieza en d (+3) a (+3) u (+3) l . No avanza más porque el final es la g.
```

### Reemplazar/sustitución caracteres.

Dado que las cadenas son objetos inmutables, lo que devuelve esta función es una **copia** de la cadena original.

```python
print("cadena larga".replace("a", "W", 3)) # cWdenW lWrga
```

### Última ocurrencia de un caracter. (equivalente a `lastIndexOf()` en Java)

```python
cadena = "Paradigmas es un grupo ejemplar"
print(len(cadena)) # 31
print(len(cadena) - cadena[::-1].find('e') - 1) # 25, cadena[::-1] invierte cadena
```

### Eliminación de todos los espacios al inicio y fin (`strip()`)

```python
print('   Hola Mundo    '.strip()) # Hello
```

### Comprobar si una cadena empieza y/o acaba por (`startswith()`, `endswith()`)

```python
print("Casa del campo".startswith("Casa")) # True
print("Casa del campo".endswith("ampa")) # False
```

### Unión de elementos.

```python
a = ['Esto', 'es', 'una', 'cadena']
print("-".join(a)) # Esto-es-una-cadena. Usamos el conector "-"
print(" ".join(a)) # Esto es una cadena. Usamos el conector " "
```

### Repetición de elementos.

```python
a = "^"
print(a*45)
```

### Expresiones regulares

```python
import re
patron = re.compile('a[2-6]+')
cadena = 'ba3425 a6' # La coincidencia del patrón no está al principio!
# findall() devuelve todas las ocurrencias en forma de lista
print("Usando findall(): ", patron.findall(cadena)) # ['a3425', 'a6']
# search() busca en la cadena alguna ocurrencia del patrón
print("Usando search(): ", patron.search(cadena)) # <_sre.SRE_Match object at 0x0311C3D8>
# match() igual que search() pero coincidencia sólo al comienzo de la cadena
print("Usando match(): ", patron.match(cadena)) # None
```
